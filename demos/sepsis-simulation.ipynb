# Sepsis Risk Score Simulation
This notebook demonstrates simplified risk scoring logic for an early sepsis detection system.

```
import numpy as np
import pandas as pd

# simulate risk scores
np.random.seed(0)
scores = np.random.gamma(2, 2, 300) / 20
df = pd.DataFrame({'risk_score': scores})

ALERT_THRESHOLD = 0.25

def classify(score):
    return 'alert' if score > ALERT_THRESHOLD else 'no_alert'

 df['decision'] = df['risk_score'].apply(classify)
df.head()
```
